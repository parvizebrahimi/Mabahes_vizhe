```python
def parse_student_number(student_number):
    # بررسی طول شماره دانشجویی
    if len(student_number) != 14:
        return "شماره دانشجویی باید ۱۴ رقم باشد."

    # استخراج اطلاعات از شماره دانشجویی
    year_of_entry = student_number[0:2]
    semester_code = student_number[2]
    degree_code = student_number[3]
    admission_code = student_number[4]
    college_code = student_number[6:9]
    major_code = student_number[8:11]

    # تعیین سال ورود
    year_of_entry = 1400 + int(year_of_entry)

    # تعیین نیم‌سال ورود
    if semester_code == '1':
        semester = "ورودی نیم‌سال اول (مهر)"
    elif semester_code == '2':
        semester = "ورودی نیم‌سال دوم (بهمن)"
    else:
        semester = "نیم‌سال ورود نامعتبر است."

    # تعیین نوع مقطع تحصیلی
    if degree_code == '1':
        degree = "کاردانی پیوسته"
    elif degree_code == '2':
        degree = "کارشناسی ناپیوسته"
    elif degree_code == '3':
        degree = "کارشناسی پیوسته"
    elif degree_code == '4':
        degree = "کاردانی ناپیوسته"
    else:
        degree = "نوع مقطع تحصیلی نامعتبر است."

    # تعیین نوبت پذیرش
    if admission_code == '0':
        admission = "شبانه"
    elif admission_code == '1':
        admission = "روزانه"
    else:
        admission = "نوبت پذیرش نامعتبر است."

    # تعیین شناسه آموزشکده
    college = ""
    if college_code == '033':
        college = "آموزشکده فنی و حرفه‌ای شهید شمسی پور تهران"
    elif college_code == '073':
        college = "آموزشکده فنی و حرفه‌ای شهید چمران کرمان"
    else:
        college = "شناسه آموزشکده نامعتبر است."

    # تعیین کد رشته تحصیلی
    major = ""
    if major_code == '302':
        major = "کامپیوتر نرم‌افزار"
    else:
        major = "کد رشته تحصیلی نامعتبر است."

    # ساخت و نمایش خروجی نهایی
    output = f"""
    سال ورود: {year_of_entry}
    نیم‌سال ورود: {semester}
    نوع مقطع تحصیلی: {degree}
    نوبت پذیرش: {admission}
    آموزشکده: {college}
    رشته تحصیلی: {major}
    """
    return output

# دریافت شماره دانشجویی از کاربر
student_number = input("شماره دانشجویی را وارد کنید: ")
print(parse_student_number(student_number))
```
